```python
import pandas as pd
import numpy as np
import re
from itertools import combinations
from difflib import get_close_matches
```

```python
from sklearn.feature_extraction.text import TfidfTransformer, CountVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
from google.colab import files

files.upload_file('reviews')
```

```python
reviews = pd.read_csv('/content/reviews', sep=',')
```

```python
reviews.name.values
```

```python
reviews[reviews['name'] == get_close_matches('Whitney ', reviews.name.values)[0]].index.str()
```

```python
reviews = reviews[['id', 'name', 'comment', 'likes', 'rating']]

reviews['comment'] = reviews['comment'].fillna('')
```

```python
dict_indices = reviews['id'].to_dict()
dict_inverse_indices = {valor: clave for clave, valor in dict_indices.items()}
```

```python
def clean_numbers(comment):
  c = comment.lower()
  c = re.sub(r'\d+', "", c)

  return c
```

```python
word_count = CountVectorizer(preprocessor=clean_numbers,
                min_df=5)
```

```python
comment_bag_of_word = (
    word_count.fit_transform(reviews['comment'])
                     .toarray()
)
```

```python
comment_bag_of_word
```

```python
columns_comment = [tup[0] for tup in sorted(word_count.vocabulary_.items(), key= lambda x: x[1])]
```

```python
df = pd.DataFrame(comment_bag_of_word, columns=columns_comment, index=reviews['name'])
```

```python
df.head()
```

```python
tf_idf = TfidfTransformer()
```

```python
tf_idf_comment = tf_idf.fit_transform(comment_bag_of_word).toarray()
```

```python
tf_idf_df = pd.DataFrame(tf_idf_comment, columns=columns_comment, index=reviews['name'])
```

```python
tf_idf_df
```

```python
similarity = cosine_similarity(tf_idf_df)
```

```python
matrix_similarity = pd.DataFrame(similarity, columns=reviews['name'], index=reviews['name'])
matrix_similarity
```

```python
orden_x_fila_similarity = np.argsort((-similarity), axis=1)
orden_similarity = np.sort(-similarity, axis=1)
```

```python
def top_k_similarities(comment_id, k):
  fila = dict_indices[comment_id]
  list_comment_sort = orden_x_fila_similarity[fila]
  list_comment = orden_similarity[fila]

  top_k = list_comment_sort[:k]
  top_k_similarities = list_comment[:k]

  top_k_df = reviews.loc[top_k].copy()
  top_k_df['similarity'] = top_k_similarities

  return top_k_df.sort_values(by='likes', ascending=False)
```

```python
def top_k_similarities_by_name(name, k):
  fila = reviews[reviews['name'] == get_close_matches(name+' ', reviews.name.values)[0]].index.tolist()[0]
  print(fila)
  list_comment_sort = orden_x_fila_similarity[fila]
  list_comment = orden_similarity[fila]

  top_k = list_comment_sort[1:k]
  top_k_similarities = list_comment[1:k]

  top_k_df = reviews.loc[top_k].copy()
  top_k_df['similarity'] = top_k_similarities

  return top_k_df.sort_values(by='likes', ascending=False)
```

```python
def top_k_similarities_by_comment(comment, k):
  fila = reviews[reviews['comment'] == get_close_matches(comment+' ', reviews.comment.values)[0]].index.tolist()[0]
  print(fila)
  list_comment_sort = orden_x_fila_similarity[fila]
  list_comment = orden_similarity[fila]

  top_k = list_comment_sort[:k]
  top_k_similarities = list_comment[:k]

  top_k_df = reviews.loc[top_k].copy()
  top_k_df['similarity'] = top_k_similarities

  return top_k_df.sort_values(by='likes', ascending=False)
```
